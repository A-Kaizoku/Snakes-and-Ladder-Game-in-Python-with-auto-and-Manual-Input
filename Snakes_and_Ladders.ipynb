````python
import random
import sys
#sorry I am using sys library beacuse i have having many errors with exit() and quit() in the player typing quit case

snakes= {17 : 7, 54 : 34, 62 : 19, 98 : 79}
ladders={3 : 38, 24 : 33, 42 : 93, 72 : 84}
next_turn=["Please proceed." ,"Lets Go.","Your turn.","Lets win this.","Are you ready?", "",]

####################################################################################################################################
def welcome_msg():
    
    msg = """##### Welcome to Snakes & Ladders Game #####"""
    print(msg)
    
####################################################################################################################################
def get_name(t):
    
    name=input("Player " + str(t)+ " name ")
    return name

####################################################################################################################################
def is_int(x,player):  #i used try and except to handle errors and to distinguish the input from string to int  
    while True:
        try:
            val=int(x)
            print(" \n Manual mode entered \n")
            return val
        
        except ValueError:
            print("\n Auto mode Entered \n ")
            if x=="roll":
                print("Rolling...")
                dice_value=random.randint(1,6)
                txt="its a {value}".format(value=dice_value)
                print(txt)
                return dice_value
            
            elif x=="quit":
                break
                
            else:
                print("this is not Valid Input")
                return 0
    print(player + " Forfeits the Game ")
    if player==player_1_name:
        print("Player 2 won")
    else:
        print("player 1 won")
    print("*"*70)
    sys.exit(0)
    
#I know this may not be the best solution but I am new to python and mujhe koi aur tareeka nahi suuj rha tha XD
    
####################################################################################################################################
def get_dice_value(player):
    while True:
        roll=input("type roll to start").strip()
        dice=is_int(roll,player)
        if dice==0:
            continue
        
        #############
        else:
            if dice>=1 and dice<21:
                return dice 
            else :
                print("\n Out of bonds \n")
                
####################################################################################################################################

def snake_ladder(player,current,dice_value):
    old=current
    current=current + dice_value
    
    if current>100:  #handling overshoot condition
        print(" \n only " + str(100 - old) + " needed to win ") 
        return old
    print("\n " + player + " moved from " + str(old) + " to " + str(current))
    print("Your Final Position : " + str(current) )
    
    if current in snakes:
        final=snakes[current]
        snake_bite(current,final,player) ################
        
    elif current in ladders :
        final=ladders[current]
        ladder(current,final,player)
    else:
        final=current
    return final


####################################################################################################################################

def snake_bite(old,current,player):
    print("\n BITTEN '`,,,''',,,''',,,''',,,```")
    print("\n" + player + " goes down to " + str(current) + " from " + str(old))
    
####################################################################################################################################

def ladder(old,current,player):
    print("\n LADDER FOUND " + "!"*50)
    print("\n" + player + " goes up to " + str(current) + " from " + str(old))
    
####################################################################################################################################

def check_win(player,position):
    if position==100 :
        print("\n "+player + " won the Game \n")
        print(" ###### Game Successfully Finished GGWP #####")
        sys.exit(0)  
    else:
        return 0
    
####################################################################################################################################

def get_type():
    print("Enter \'1\' for auto mode \n or \n \'2\' for manual mode ")
    x=int(input())
    return x
    
####################################################################################################################################    
def start():
    welcome_msg()
    
    #fetching names
    global player_1_name,player_2_name
    
    player_1_name=get_name(1)
    player_2_name=get_name(2)
    print("\n ###### Let us Start ##### \n")
    
    p1_initial_position=0
    p2_initial_position=0
    
    
    
    while True:    #while True means loop runs forever
        
        print( "\n" + player_1_name + " " + random.choice(next_turn))
        dice_value=get_dice_value(player_1_name)
        print(" you got "+ str(dice_value))
        print(player_1_name + " moving...")
        
        p1_initial_position=snake_ladder(player_1_name,p1_initial_position,dice_value)
        check_win(player_1_name,p1_initial_position)
        
        #####################################################################
       
        print( "\n\n" + player_2_name + " " + random.choice(next_turn))
        dice_value=get_dice_value(player_2_name)
        print(" you got "+ str(dice_value))
        print(player_2_name + " moving...")
        
        p2_initial_position=snake_ladder(player_2_name,p2_initial_position,dice_value)
        check_win(player_2_name,p2_initial_position)
####################################################################################################################################
    
start()
 
   
````
